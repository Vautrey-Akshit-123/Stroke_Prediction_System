```python
%matplotlib inline
from load_dataset_module import DatasetLoader
from eda_module import DataExplorer
from ui_module import UserInterface
import sys
from contextlib import redirect_stdout
import os

# Redirect preprocessing and EDA output to a log file
log_file = os.path.join(os.getcwd(), "preprocessing_log.txt")
try:
    with open(log_file, 'w') as f:
        with redirect_stdout(f):
            # Background: Generate preprocessed_data.csv
            loader = DatasetLoader("data.csv")
            loader.load_data()
            loader.handle_missing_values()
            categorical_cols = ["Gender", "Work Type", "Residence Type", "Smoking Status", 
                               "Physical Activity", "Dietary Habits", "Alcohol Consumption", 
                               "Education Level", "Income Level", "Region"]
            loader.encode_categorical(categorical_cols)
            loader.save_preprocessed_data("preprocessed_data.csv")

            # Background: Generate featured_data.csv
            explorer = DataExplorer("preprocessed_data.csv")
            explorer.handle_imbalance("Stroke Occurrence")
            explorer.compute_features()
except Exception as e:
    print(f"Error during preprocessing or EDA: {e}")
    raise

# Run the user interface (output in notebook)
ui = UserInterface()
ui.run_ui()
```
